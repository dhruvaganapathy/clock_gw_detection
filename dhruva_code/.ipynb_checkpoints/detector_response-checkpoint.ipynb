```python
d = 1e10
c = 3e8


def h(t,H,f,phi):
    return H*np.sin(2*np.pi*f*t+phi)

def gw(t0,T):
    ts = np.linspace(t0,t0+T,10000)
    s = 1/2*(h(ts)-h(ts+d/c))
    return (ts,s) 


def signal(t0,T=160,H= 1e-20,f = 1/160,phi = 0, theta = np.pi/2, pol = 0):
    
    ts = np.linspace(t0,t0+T,10000)
    dt = ts[1]-ts[0]
    s = (1+np.cos(theta))/2*(h(ts,H,f,phi)-h(ts+d*(1-np.cos(theta))/c,H,f,phi))
    s_bar = np.sum((1/T * s*dt))
    return s_bar

```

```python
theta = np.linspace(-np.pi,np.pi,100)
ant_p = []
for t in theta:
    ant_p.append((np.abs(signal(0,theta = t))))
```

```python
plt.plot(theta,ant_p/max(ant_p))
```

```python

fs = np.linspace(1/320,0.1,500)
ts = np.linspace(0,1000,500)
signal_max = []
for f in fs: 
    signals = []
    for t in ts:
        signal_amp = signal(t,f = f,T = 5)
        signals.append(signal_amp)
    # plt.plot(ts,signals)
    signal_max.append(max(signals))
```

```python
plt.plot(fs*1000,signal_max/max(signal_max))
plt.ylabel('Detector Response @ T = 5s')
plt.xlabel('Frequency [mH')
```

```python
phi = np.linspace(0, 2*np.pi, 100)
theta = np.linspace(0, np.pi, 100)
Phi, Theta = np.meshgrid(phi, theta)  # Create 2D grid

# Compute F_plus and F_cross
F_plus = np.sin(Phi)**2 - np.cos(Phi)**2 * np.cos(Theta)**2
F_cross = 2*np.sin(Phi) * np.cos(Phi) * np.cos(Theta)

F_rms = np.sqrt(F_plus**2+F_cross**2)
# Plot F_plus
plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.pcolormesh(Phi, Theta, F_plus, shading='auto', cmap='viridis')
plt.colorbar(label='F_plus')
plt.xlabel('phi [rad]')
plt.ylabel('theta [rad]')
plt.title('F_plus(, )')

# Plot F_cross
plt.subplot(1,3,2)
plt.pcolormesh(Phi, Theta, F_cross, shading='auto', cmap='plasma')
plt.colorbar(label='F_cross')
plt.xlabel('phi [rad]')
plt.ylabel('theta [rad]')
plt.title('F_cross(, )')


plt.subplot(1,3,3)
plt.pcolormesh(Phi, Theta, F_rms, shading='auto', cmap='plasma')
plt.colorbar(label='F_rms')
plt.xlabel('phi [rad]')
plt.ylabel('theta [rad]')
plt.title('F_rms(, )')

plt.tight_layout()
```

```python
def spherical_to_cartesian(F):
    X = F * np.sin(Theta) * np.cos(Phi)
    Y = F * np.sin(Theta) * np.sin(Phi)
    Z = F * np.cos(Theta)
    return X, Y, Z

X_plus, Y_plus, Z_plus = spherical_to_cartesian(F_plus)
X_cross, Y_cross, Z_cross = spherical_to_cartesian(F_cross)
X_rms, Y_rms, Z_rms = spherical_to_cartesian(F_rms)

# Plot F_plus
fig = plt.figure(figsize=(12,5))
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X_plus, Y_plus, Z_plus, facecolors=plt.cm.viridis((F_plus - F_plus.min()) / (F_plus.max() - F_plus.min())),
                 rstride=1, cstride=1, antialiased=True)
ax1.set_title('F_plus 3D')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# Plot F_cross
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(X_cross, Y_cross, Z_cross, facecolors=plt.cm.plasma((F_cross - F_cross.min()) / (F_cross.max() - F_cross.min())),
                 rstride=1, cstride=1, antialiased=True)
ax2.set_title('F_cross 3D')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

ax2 = fig.add_subplot(133, projection='3d')
ax2.plot_surface(X_rms, Y_rms, Z_rms, facecolors=plt.cm.plasma((F_rms- F_rms.min()) / (F_rms.max() - F_rms.min())),
                 rstride=1, cstride=1, antialiased=True)
ax2.set_title('F_rms 3D')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')



```
